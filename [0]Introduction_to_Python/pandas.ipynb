### Pandas Series Object

The pandas series can be created from a list or array. The constructor is ``pd.Series()``. In the example below we pass a list ``[0.25,.5,.75,1.0]``.

```
data = pd.Series([0.25,.5,.75,1.0])
data
```

```
data.values
```

```
data.index
```

```
## indexing works the same as numpy arrays
data[1]
```

```
data[1:3]
```

```
## indices can be different
data = pd.Series([.25, .5, .75, 1], index = ['a', 'b', 'c', 'd'])
```

```
data.values
```

```
data.index
```

```
## It works as a dictionary as well
data['a']
```

### Pandas DataFrame Object

If a Series is an analog of a one-dimensional array with flexible indices, a DataFrame is an analog of a two-dimensional array with both flexible row indices and flexible column names.

```
area = pd.Series({'California': 423967, 
                  'Texas': 695662,
                  'New York': 141297, 
                  'Florida': 170312,
                  'Illinois': 149995})
```

```
pop = pd.Series({'California': 38332521, 
                 'Texas': 26448193,
                 'New York': 19651127, 
                 'Florida': 19552860,
                 'Illinois': 12882135})
```

```
## We can combine two Series with common indices
data = pd.DataFrame({'area':area, 'pop':pop})
```

```
data
```

```
data.values
```

```
data.index
```

```
## we can select variables using the variable name in brackets and quotes
data['area']
```

```
## We can also select the variable using a ., as an attribute
data.area
```

```
## Creating new variables is easy
data['density'] = data['pop'] / data['area']
```

```
data
```

```
## Matrix operations area also possible
## e.g. Transpose
data.T
```

```
## You can remove exponential notation by using the .set_option() method
pd.set_option('display.float_format', lambda x: '%.5f' % x)
data.T
```

### Pandas DataFrame Indexing

For indexing, Pandas DataFrames use the ``loc`` and ``iloc`` indexers. 

The ``loc`` attribute allows indexing and slicing that always references the explicit index:

```
data.loc['California']
```

The iloc attribute allows indexing and slicing that always references the implicit Python-style index:

```
## first row
data.iloc[0,:]
```
