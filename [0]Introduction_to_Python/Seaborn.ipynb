Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. The usual alias for seaborn is `sns`.

```
import seaborn as sns
```

```
## Let's import some other modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
# let's load a sample dataset from the Seaborn module
tips = sns.load_dataset('tips')
tips
```

### Scatter Plot

```
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```
## color by smoker
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
```

```
## The joinplot includes histograms on the sides
sns.jointplot(x = tips['total_bill'], y = tips['tip'])
```

```
## the regplot creates a regression line with the scatterplot
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
```

```
## the lmplot also does the same
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

### Line Plot

```
## let's create a dataset. Make sure you understand the code.
df = pd.DataFrame(dict(time = np.arange(500), value = np.random.randn(500).cumsum()))
```

```
df.head(10)
```

```
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df)
```

### Categorical Plots


```
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```
## There are interesting variations using the kind and hue options
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm', hue = 'size')
```

### Box Plots

You can use the `kind = "box"` for box plots

```
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips)
```

```
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips, hue = 'sex')
```

### Scatter Plot Matrix

```
## You can pass the whole dataset to the pairplot function and it will create a scatterplot matrix of the numeric variables
sns.pairplot(tips)
```
