```python
# prompt: graph the function J(w0, w1) = w0^2 + w1^2 - 4*w0 - 4*w1 + 3

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function J(w0, w1)
def J(w0, w1):
    return w0**2 + w1**2 - 4*w0 - 4*w1 + 3

# Create a grid of w0 and w1 values
w0_vals = np.linspace(-2, 6, 100)
w1_vals = np.linspace(-2, 6, 100)
w0, w1 = np.meshgrid(w0_vals, w1_vals)

# Calculate the corresponding J(w0, w1) values
J_vals = J(w0, w1)

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(w0, w1, J_vals)

ax.set_xlabel("w0")
ax.set_ylabel("w1")
ax.set_zlabel("J(w0, w1)")
ax.set_title("Plot of J(w0, w1) = w0^2 + w1^2 - 4w0 - 4w1 + 3")

plt.show()
```

```python
## gradient
def gradient(w0, w1):
  return np.array([2*w0 - 4, 2*w1 - 4])
```

```python
## Gradient Descent Algorithm
## Gradient (derivatives)
## intial_guess : usually like a random normal but if you have a better guess (hot start)
## learning_rate : alpha, step size. Usually small 0.01
## max_iter: the highest number of steps
## tolerance: how close the gradient to zero. STOP if the model gradiet is less than that

def gradient(w):
  return 2*w - 4

def gradient_descent(initial_guess = np.random.randn(), learning_rate = 1, max_iter = 100, tolerance = 0.0001):
  w = initial_guess

  for i in range(max_iter):
    grad = gradient(w) ## derivative of J(w)
    if np.linalg.norm(grad) < tolerance:
      break
    w = w - learning_rate*grad
    print(f"Iteration {i+1}: w = {w}, J(w) = {grad}")
  return w
```

```python
gradient_descent()
```

```python
## learning_rate: 0.001 takes too long to get to the optimal
## there is a point where dJ(w) bouncing from positive to negative
```

```python
## Tensorflow NN
## w = y-intercept and slope
## w = [w0, w1]
import tensorflow as tf

## reshape (column vector)
y = y.reshape(-1,1)  ## -1 means just estimate the number of rows

## model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape = (1,)),      ## input layer
    tf.keras.layers.Dense(1)                  ## output layer
])

## compile
## optimizer =
model.compile(optimizer='adam', loss='mse')
```

```python
## how many parameters w you are estimating
model.summary()
```

```python
## fit
## epochs = 100 the gradient descent is going to pass over the whole data set 100 times
## we will update w every 10 obs
model.fit(x.reshape(-1,1), y.reshape(-1,1), epochs=100, batch_size = 10)
```
