## **Matrices**

```python
import numpy as np


M1 = np.array(
    [
        [2, -5, -1, 0],
        [-9, 4, 6, 13],
        [4, 5, 12, -2]
    ]
)
```

```python
## type
type(M1)
```

```python
## shape
M1.shape
```

```python
## M2 (4 x 2)

M2 = np.array(
    [
        [1,2],
        [3,4],
        [5,6],
        [7,8]
    ]
)
```

```python
## multiply
M1.dot(M2)
```

```python
##
M3 = M1 @ M2
```

```python
np.linalg.matrix_rank(M3)
```
