## **The Gradient Descent Algorithm**

Optimization technique to achieve a minimum value of a cost function.

```python
## Suppose we have cost function Jw = w**2 - 4w + 3
## Assume for now we have one single weight w

import numpy as np
import matplotlib.pyplot as plt

w = np.linspace(-2,6,100)

## Create the cost function
def J(w):
  return w**2 - 4*w + 3

## plot the curve
plt.plot(w, J(w));
```

```python
## Gradient Function
#### It will give the SLOPE of each weight.

def dJ(w):
  return 2*w - 4
```

```python
## e.g., what w = 4
dJ(4)   ## slope is positive (NOT OPTIMAL)
```

**(NUMBER 1): Down the gradient**

We want to move towards $-\nabla J(w)$

```python
-dJ(2)
```

**(NUMBER 2): Step size (Learning rate)**

We don't want to give a step of the size of $-\nabla J(w)$, because it is generally too large. We give a step size of $-\alpha \nabla J(w)$. $\alpha$ is the learning rate (step size). Sometimes it is also called $\eta$

```python
## learning rate = 0.1
w = 1
-0.1*dJ(w)
```

```python
## learning rate = 0.05
## Suppose w = 3
## figure out by how much w has to change (-.1)
w = 3
-0.05*dJ(3)

## the new w will be 3-0.1 = 2.9
```
